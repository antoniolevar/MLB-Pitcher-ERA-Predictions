## MLB ERA Predictions

### ERA (Earned Run Average): The average number of earned runs a pitcher allows per 9 innings. ((ER*9)/IP)

```python
import os
import pandas as pd
import numpy as np
from pybaseball import pitching_stats
```

```python
START = 2012
END = 2023
```

```python
pitching_stats = pitching_stats(START, END, qual=200)
pitching_stats
```

```python
pitching_stats = pitching_stats.groupby('IDfg', group_keys=False).filter(lambda x: x.shape[0] > 1)
```

```python
def next_season(player):
    player = player.sort_values('Season')
    player['Next_ERA'] = player['ERA'].shift(-1)
    return player

pitching_stats = pitching_stats.groupby('IDfg', group_keys = False).apply(next_season)
```

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = Ridge(alpha=1)

split = TimeSeriesSplit(n_splits = 3)

sfs = SequentialFeatureSelector(rr, n_features_to_select = 20, direction = 'forward', cv = split)
```

```python
removed_columns = ["Next_ERA", "Name", "Team", "Season", "IDfg"]
selected_columns = pitching_stats.columns[~pitching_stats.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
pitching_stats.loc[:,selected_columns] = scaler.fit_transform(pitching_stats[selected_columns])
```

```python
pitching_stats.describe()
```

```python
pitching_stats.dropna(subset=["Next_ERA"], inplace=True)
sfs.fit(pitching_stats[selected_columns], pitching_stats["Next_ERA"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```

```python
def backtest (data, model, predictors, start = 5, step = 1):
    all_predictions = []
    
    years = sorted(data["Season"].unique())
    for i in range(start, len(years), step):
        current_year = years[i]
        
        train = data[data["Season"] < current_year]
        test = data[data["Season"] == current_year]
        
        model.fit(train[predictors], train["Next_ERA"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test['Next_ERA'], preds], axis=1)
        combined.columns = ['Actual', 'Predicted']
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(pitching_stats, rr, predictors)
predictions
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(predictions["Actual"], predictions["Predicted"])
```

```python
def player_history(df):
    df = df.sort_values('Season')
    
    df['Player_Season'] = range(0, df.shape[0])
    df['ERA_Corr'] = list(df[['Player_Season', 'ERA']].expanding().corr().loc[(slice(None), 'Player_Season'), 'ERA'])
    # Filling n/a with '1' implies a 1-to-1 correlation.
    df['ERA_Corr'].fillna(1, inplace = True)
    
    df['ERA_Diff'] = df['ERA'] / df['ERA'].shift(1)
    df['ERA_Diff'].fillna(1, inplace = True)
    
    df['ERA_Diff'][df['ERA_Diff'] == np.inf] = 1
    
    return df
pitching_stats = pitching_stats.groupby('IDfg', group_keys = False).apply(player_history)
```

### New Column Meanings
* Player Season - The season of which that player is in during their major league career.
* ERA Corr - Evaluates the player's ERA for any correlation between the season and ERA.
* ERA Diff - Evaluates the difference/ratio in ERA from the previous year to the current year.

```python
def group_averages(df):
    return df['ERA'] / df['ERA'].mean()
```

```python
pitching_stats['ERA_Season'] = pitching_stats.groupby('Season', group_keys = False).apply(group_averages)
new_predictors = predictors + ['Player_Season','ERA_Corr', 'ERA_Diff', 'ERA_Season']
predictions = backtest(pitching_stats, rr, new_predictors)
mean_squared_error(predictions["Actual"], predictions["Predicted"])
```

## Takeaways:
* IBB (Intentional Walks): Number of intentional walks issue by the pitcher.

* E-F (ERA-FIP Differential): The difference between the pitcher’s ERA and FIP.
    - FIP (Fielding Independent Pitching): An estimate of a pitcher’s ERA based on strikeouts, walks/HBP, and home runs allowed, assuming league average results on balls in play.
    - ERA (Earned Run Average): The average number of earned runs a pitcher allows per 9 innings. ((ER*9)/IP).

* Z-Swing% (Inside the Zone Swing Percentage): Swings at pitches inside the zone divided by pitches inside the zone.

* Framing (FRM): According to mlb.com, catcher framing is the art of a catcher receiving a pitch in a way that makes it more likely for an umpire to call it a strike -- whether that's turning a borderline ball into a strike, or not losing a strike to a ball due to poor framing. The effects of a single pitch can be huge; in 2019, hitters had an .858 OPS after a 1-0 count, but just a .631 OPS after an 0-1 count.

* LOB% (Left On-Base Percentage): Percentage of pitcher’s own base runners that they strand over the course of a season. Not equal to the LOB column in the box score.


According to the result table these are the significant factors at the 5 or 10% level.
